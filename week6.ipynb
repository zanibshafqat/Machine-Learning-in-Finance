```python
import pandas as pd
import plotly.graph_objs as go

# Read price data
csv_file = "XAUUSD_1 Min_Bid_2022.01.01_2022.12.31.csv"
df = pd.read_csv(csv_file)

# SID-based indexing
sid_num = 2425274
start_idx = int(str(sid_num)[-5:])
window_size = int(str(sid_num)[-3:])

# Slice required rows
price_slice = df.iloc[start_idx : start_idx + window_size]

# Plot traces
high_line = go.Scatter(
    x=price_slice.index,
    y=price_slice["High"],
    mode="lines",
    name="High"
)

low_line = go.Scatter(
    x=price_slice.index,
    y=price_slice["Low"],
    mode="lines",
    name="Low"
)

# Layout settings
chart_layout = go.Layout(
    title="Gold Price High–Low Movement",
    xaxis=dict(title="Time (Minutes)"),
    yaxis=dict(title="Price"),
    template="plotly"
)

# Render figure
figure = go.Figure(data=[high_line, low_line], layout=chart_layout)
figure.show()
```
