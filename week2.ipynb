```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('adult_data_mini.csv')
```

```python
data.head()
```

```python
def adjust_hours(row, n):
    if pd.notna(row['hours-per-week']):
        return row['hours-per-week'] - n
    return row['hours-per-week']  # Leave NaN values as they are

# Apply the function to the dataframe
n = 4  # My calculated calculated n
data['reduced-hours-per-week'] = data.apply(adjust_hours, axis=1, n=n)
```

```python
grouped_original = data.groupby(['relationship', 'hours-per-week']).size()
print("Grouped by 'relationship' and 'hours-per-week':")
print(grouped_original)
```

```python
grouped_reduced = data.groupby(['relationship', 'reduced-hours-per-week']).size()
print("Grouped by 'relationship' and reduced 'hours-per-week':")
print(grouped_reduced)
```
