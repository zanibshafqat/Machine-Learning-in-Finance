```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense

# Read CSV data
df = pd.read_csv("data_stocks.csv")

# Separate inputs and output
features = df.iloc[:, :-1].to_numpy()
target = df.iloc[:, -1].to_numpy()

# Train–test split
X_tr, X_te, y_tr, y_te = train_test_split(
    features, target, test_size=0.2, random_state=42
)

# Feature normalization
scaler = StandardScaler()
X_tr = scaler.fit_transform(X_tr)
X_te = scaler.transform(X_te)

# Hidden layer size derived from SID
sid_value = 274
neurons_l1 = sid_value
neurons_l2 = neurons_l1 // 2

# Build MLP model
mlp_model = Sequential()
mlp_model.add(Dense(neurons_l1, activation="relu", input_shape=(X_tr.shape[1],)))
mlp_model.add(Dense(neurons_l2, activation="relu"))
mlp_model.add(Dense(1))

# Model configuration
mlp_model.compile(
    optimizer="adam",
    loss="mean_absolute_error",
    metrics=["mae"]
)

# Training phase
mlp_model.fit(
    X_tr,
    y_tr,
    epochs=10,
    batch_size=32,
    validation_data=(X_te, y_te),
    verbose=1
)

# Performance evaluation
test_mae = mlp_model.evaluate(X_te, y_te, verbose=0)[1]
print(f"Custom MLP MAE: {test_mae:.4f}")

# Network structure
mlp_model.summary()
```
