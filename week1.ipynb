```python
import numpy as np

vector = np.arange(74)
print("Vector:", vector)
```

```python

matrix = vector.reshape(1, -1)
print("2-D Array with 1 Row:")
print(matrix)
```

```python
new_array = matrix.copy()
print("Copied Array:")
print(new_array)
```

```python
print("Shape of the Array:", new_array.shape)
```
