```python
import numpy as np

vector = np.arange(67)
print("Vector:", vector)

matrix = vector.reshape(1, -1)  # Reshape to 1 row and as many columns as needed
print("2-D Array with 1 Row:")
print(matrix)

new_array = matrix.copy()
print("Copied Array:")
print(new_array)

print("Shape of the Array:", new_array.shape)
```
